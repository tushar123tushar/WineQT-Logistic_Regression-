```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_set = pd.read_csv('WineQT.csv')
```

```python
data_set
```

```python
sns.heatmap(data_set.isnull() , yticklabels=False , cbar=False)
```

```python
x = data_set.drop('quality' , axis=1)
y = data_set['quality']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

logistic = LogisticRegression()
```

```python
parameters  = {'penalty': ['elasticnet', 'l2', 'l1'] , 'C' : [1,2,3,4,5,6,10,5,450,102] , 'max_iter':[100,200,300]}
```

```python
class_fire = GridSearchCV(logistic , param_grid=parameters , scoring='accuracy' , cv=5)
class_fire
```

```python
class_fire.fit(X_train , y_train)
```

```python
print(class_fire.best_params_)
```

```python
print(class_fire.best_score_)
```

```python
from sklearn.metrics import r2_score , accuracy_score , classification_report
x_pred = class_fire.predict(X_test)
```

```python
accuracy_score(x_pred , y_test)
```

```python
classification_report(x_pred , y_test)
```
